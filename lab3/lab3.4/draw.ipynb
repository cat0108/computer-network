```python
import matplotlib.pyplot as plt
# 使用宋体显示中文
plt.rcParams['font.sans-serif'] =   ['SimSun']

# 数据
packet_loss_rate = [2, 4, 6, 8, 10]
stop_and_wait = [24558, 30056, 35935, 41897, 46669]
gbn = [31275, 58456, 123222, 328037, 591808]
sr = [16886, 17480, 18317, 21530, 28005]

# 绘制图表
plt.figure(figsize=(10, 6))

plt.plot(packet_loss_rate, stop_and_wait, marker='o', label='停等机制')
plt.plot(packet_loss_rate, gbn, marker='o', label='滑动窗机制(GBN)')
plt.plot(packet_loss_rate, sr, marker='o', label='滑动窗机制(SR)')

# 设置图表属性
plt.title('不同丢包率下不同机制的传播时延比较')
plt.xlabel('丢包率 (%)')
plt.ylabel('时延:ms (对数处理)')
plt.legend()

# 对纵坐标进行对数处理
plt.yscale('log')

# 关闭网格线
plt.grid(False)

# 显示图表
plt.show()
```
