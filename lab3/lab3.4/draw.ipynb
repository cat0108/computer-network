```python
import matplotlib.pyplot as plt
# 
plt.rcParams['font.sans-serif'] =   ['SimSun']

# 
packet_loss_rate = [2, 4, 6, 8, 10]
stop_and_wait = [24558, 30056, 35935, 41897, 46669]
gbn = [31275, 58456, 123222, 328037, 591808]
sr = [16886, 17480, 18317, 21530, 28005]

# 
plt.figure(figsize=(10, 6))

plt.plot(packet_loss_rate, stop_and_wait, marker='o', label='')
plt.plot(packet_loss_rate, gbn, marker='o', label='(GBN)')
plt.plot(packet_loss_rate, sr, marker='o', label='(SR)')

# 
plt.title('')
plt.xlabel(' (%)')
plt.ylabel(':ms ()')
plt.legend()

# 
plt.yscale('log')

# 
plt.grid(False)

# 
plt.show()
```
